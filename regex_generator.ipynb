```python
import pandas as pd
```

```python
df = pd.read_csv("verbose_text", sep=" - ", )
```

```python
df["hard_regex"] = "/\\b" + df["hard"] + "\\b/gi"
```

```python
df["escaped_easy"] = "\"" + df["easy"] + "\""
```

```python
print("[")
for i in range(len(df)):
    print(f"[{df.hard_regex[i]}, {df.escaped_easy[i]}],")
print("]")
```

```python
df["easy_regex"] = "/\\b" + df["easy"] + "\\b/gi"
df["escaped_hard"] = "\"" + df["hard"] + "\""
```

```python
print("[")
for i in range(len(df)):
    print(f"[{df.easy_regex[i]}, {df.escaped_hard[i]}],")
print("]")
```
